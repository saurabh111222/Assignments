```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
df = pd.DataFrame({
'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
    
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],

'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],

'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']
})
```

```python
for i in range(len(df['FlightNumber'])):
    if pd.isnull(df.loc[i,'FlightNumber'])==True:
        df.loc[i,'FlightNumber']=df.loc[i-1,'FlightNumber']+10
df
```

```python
lst=[]
lst1=[]
lst2=[]
for i in range(len(df['From_To'])):
    lst=lst+df['From_To'][i].split('_')
for i in range(len(lst)):
    if i%2==0:
        lst1.append(lst[i])
    else:
        lst2.append(lst[i])
```

```python
df['From']=lst1
df['To']=lst2
df.drop('From_To', axis=1)
df
```

```python
for i in range(len(df['From'])):
    df.loc[i, 'From']=df['From'][i][0].upper()+df['From'][i][1:].lower()
    df.loc[i, 'To']=df['To'][i][0].upper()+df['To'][i][1:].lower()
df
```

```python
df1=pd.DataFrame(df['RecentDelays'].values.tolist(), columns=['Delay_1','Delay_2','Delay_3'])
df3=pd.concat([df1,df], axis=1)
df3
```
